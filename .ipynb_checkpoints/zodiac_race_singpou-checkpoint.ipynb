## Augment dataset with fourth positions for each race for all animals outside of top 3

```python
augment = True
```

```python
animal_list = ['rat', 'ox', 'tiger', 'rabbit', 'dragon', 'snake', 
               'horse', 'goat', 'monkey', 'rooster', 'dog', 'pig']
other_places = ['Four', 'Five', 'Six', 'Seven',
                'Eight', 'Nine','Ten', 'Eleven', 'Twelve']
```

```python
def augment_results(row):
    top3 = [row['Winner'], row['Second'], row['Third']]
    remaining = list(set(animal_list) - set(top3))
    
    for counter, place in enumerate(other_places):
        row[place] = remaining[counter]
    row['Four+_time'] = row['Third_Time'] * 1.1
    return row
```

```python
if augment:
    z_df = z_df.apply(augment_results, axis=1)
```

## Scale the numerical features - for linear models

```python
# normalize == True means subtracting the mean and dividing by std dev; normalize == False means
# scaling each features to be within 0 and 1.
normalize = False
need_to_scale = False
```

### Scale for train

```python
num_features_col_names = ['Desert', 'Mountain', 'Water', 'Land', 'Forest', 'Snow', 
                          'Average_Temperature', 'Total_dist']
```

```python
X_train_num_features = train[num_features_col_names]
```

```python
num_fea_mean_dict = {}
num_fea_std_dict = {}
train_all_num_fea_mean_dict = {}
train_all_num_fea_std_dict = {}
num_fea_min_dict = {}
num_fea_max_dict = {}
train_all_num_fea_min_dict = {}
train_all_num_fea_max_dict = {}

def scale_train_fea(col, train_all=False, normalize=True):
    
    if normalize:
        mean = np.mean(col)
        std = np.std(col)

        if train_all == False:
            num_fea_mean_dict[col.name] = mean
            num_fea_std_dict[col.name] = std
        else:
            train_all_num_fea_mean_dict[col.name] = mean
            train_all_num_fea_std_dict[col.name] = std

        return (col - mean) / std
    else:
        smallest = np.min(col)
        largest = np.max(col)
        
        if train_all == False:
            num_fea_min_dict[col.name] = smallest
            num_fea_max_dict[col.name] = largest
        else:
            train_all_num_fea_min_dict[col.name] = smallest
            train_all_num_fea_max_dict[col.name] = largest

        return (col - smallest) / (largest - smallest)
        
```

```python
X_train_num_features_scaled = X_train_num_features.apply(lambda col: 
                                                         scale_train_fea(col, 
                                                                         train_all=False, 
                                                                         normalize=normalize), axis=0)
```

### Scale for test

```python
X_test_num_features = test[num_features_col_names]
```

```python
def scale_test_fea(col, train_all=False, normalize=True):

    if normalize:    
        if train_all == False:    
            mean = num_fea_mean_dict[col.name]
            std = num_fea_std_dict[col.name]
        else:
            mean = train_all_num_fea_mean_dict[col.name]
            std = train_all_num_fea_std_dict[col.name]

        return (col - mean) / std
    else:
        if train_all == False:    
            smallest = num_fea_min_dict[col.name]
            largest = num_fea_max_dict[col.name]
        else:
            smallest = train_all_num_fea_min_dict[col.name]
            largest = train_all_num_fea_max_dict[col.name]

        return (col - smallest) / (largest - smallest)
```

```python
X_test_num_features_scaled = X_test_num_features.apply(lambda col: 
                                                       scale_test_fea(col, 
                                                                      train_all=False, 
                                                                      normalize=normalize), axis=0)
```

## Create label (y)

```python
X_train_df_dict = {}
y_train_df_dict = {}
X_test_df_dict = {}
y_test_df_dict = {}
X_train_all_df_dict = {}
y_train_all_df_dict = {}
```

```python
def extract_rank_each_animal(source_df, X_df_dict, y_df_dict, df="train"):
    
    for animal in animal_list:
        first_df = source_df[source_df['Winner'] == animal]
        cols_to_keep = num_features_col_names.copy()
        cols_to_keep.append('Winner')
        cols_to_keep.append('Winner_Time')
        first_df = first_df[cols_to_keep]
        first_df.rename(columns={'Winner':'Animal'}, inplace=True)
        first_df.rename(columns={'Winner_Time':'Time'}, inplace=True)

        second_df = source_df[source_df['Second'] == animal]
        cols_to_keep = num_features_col_names.copy()
        cols_to_keep.append('Second')
        cols_to_keep.append('Second_Time')
        second_df = second_df[cols_to_keep]
        second_df.rename(columns={'Second':'Animal'}, inplace=True)
        second_df.rename(columns={'Second_Time':'Time'}, inplace=True)    

        third_df = source_df[source_df['Third'] == animal]
        cols_to_keep = num_features_col_names.copy()    
        cols_to_keep.append('Third')
        cols_to_keep.append('Third_Time')
        third_df = third_df[cols_to_keep]
        third_df.rename(columns={'Third':'Animal'}, inplace=True)
        third_df.rename(columns={'Third_Time':'Time'}, inplace=True)    

        if augment and (df == "train" or df == "train_all"):
            for counter, place in enumerate(other_places):

                other_df = source_df[source_df[place] == animal]
                cols_to_keep = num_features_col_names.copy()
                cols_to_keep.append(place)
                cols_to_keep.append('Four+_time')                        
                other_df = other_df[cols_to_keep]
                other_df.rename(columns={place:'Animal'}, inplace=True)
                other_df.rename(columns={'Four+_time':'Time'}, inplace=True)
        #         other_df['Time'] *= max(0.8, np.random.random()) + 2

                if counter == 0:
                    others_df = other_df.copy()
                else:
                    others_df = pd.concat([others_df, other_df], axis=0)

            all_df = pd.concat([first_df, second_df, third_df, others_df], axis=0)
        else:
            all_df = pd.concat([first_df, second_df, third_df], axis=0)

        X_df_dict[animal] = all_df.drop(['Animal', 'Time'], axis=1)
        y_df_dict[animal] = all_df['Time']

        if animal == animal_list[0]:
            X_df = all_df.drop(['Time'], axis=1).copy()
            y_df = all_df['Time'].copy()
        else:
            X_df = X_df.append(all_df.drop(['Time'], axis=1))
            y_df = y_df.append(all_df['Time'])
            
    return X_df, y_df
```

```python
X_train_df, y_train_df = extract_rank_each_animal(train_scaled, X_train_df_dict, y_train_df_dict, df="train")
X_test_df, y_test_df = extract_rank_each_animal(test_scaled, X_test_df_dict, y_test_df_dict, df="test")
X_train_all_df, y_train_all_df = extract_rank_each_animal(train_all_scaled, X_train_all_df_dict, y_train_all_df_dict, df="train_all")
```
